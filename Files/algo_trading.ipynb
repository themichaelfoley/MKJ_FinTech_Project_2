```python
%matplotlib inline
import os
import requests
import pandas as pd
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi
import matplotlib.pyplot as plt
import hvplot.pandas
import numpy as np
import seaborn as sns
from pathlib import Path
```

```python
# Define function to create Pickle File
def get_api_dataframe():
    # Set Alpaca API key and secret

    load_dotenv()
    api_key= os.getenv("ALPACA_API_KEY")
    sec_key= os.getenv("ALPACA_SECRET_KEY")

# Create the Alpaca API object

    alpaca = tradeapi.REST(
        api_key,
        sec_key,
        api_version="v2"
    )
# Get 3 years worth of historical data for Alcohol stocks, Cannabis Stocks, and S&P 500

    start_date = pd.Timestamp('2019-04-20', tz='America/New_York').isoformat()
    end_date = pd.Timestamp('2022-04-20', tz='America/New_York').isoformat()

    tickers = ["BUD", "DEO", "SAM", "TAP", "STZ","VFF","JAZZ","CRON", "TLRY","CGC", "SPY"]

# Set timeframe to "1Day" for Alpaca API
    timeframe = "1Day"

# Get current closing prices for All Stocks

    df_stock_data = alpaca.get_bars(
        tickers,
        timeframe,
        start = start_date,
        end = end_date
    ).df

#Set index to date
    df_stock_data.index = df_stock_data.index.date
    return df_stock_data
```

```python
# Check if file exists
# If file does not exist then load api and save file
# else read data from pikl

api_data_file = Path("api_data.pkl")
if api_data_file.is_file():
    print ("API data file exists")
    df_stock_data = pd.read_pickle(api_data_file)
else:
    print ("API file does not exist") 
    df_stock_data = get_api_dataframe()
    df_stock_data.to_pickle(api_data_file)

#View dataframe
df_stock_data.head(-5)
```

```python
# Calculate the daily returns using the closing prices and the pct_change function
df_stock_data["actual_returns"] = df_stock_data["close"].pct_change()

# Display sample data
df_stock_data.head()
```

```python
# Drop all NaN values from the DataFrame
df_stock_data = df_stock_data.dropna()

# Review the DataFrame
display(df_stock_data.head())
display(df_stock_data.tail())
```

```python
#Create Feature Set and Window Sizes
# Define a window size of 4
short_window = 4

# Create an SMA that uses short_window, and assign it to a new column named “sma_fast”
df_stock_data["sma_fast"] = df_stock_data["close"].rolling(window=short_window).mean()


# Define a window size of 20
long_window = 20

# Create an SMA that uses long_window, and assign it to a new columns named “sma_slow”
df_stock_data["sma_slow"] = df_stock_data["close"].rolling(window=long_window).mean()
```

```python
# Drop the NaNs using dropna()
df_stock_data = df_stock_data.dropna()
```

```python
# Assign a copy of the `sma_fast` and `sma_slow` columns to a new DataFrame called `X`
X = df_stock_data[["sma_fast", "sma_slow"]].shift().dropna().copy()

# Display sample data
display(X.head())
display(X.tail())
```

```python
#Create Target Set
# Create a new column in the `df_stock_data` called "signal" setting its value to zero.
df_stock_data["signal"] = 0.0

# Create the signal to buy
df_stock_data.loc[(df_stock_data["actual_returns"] >= 0), "signal"] = 1

# Create the signal to sell
df_stock_data.loc[(df_stock_data["actual_returns"] < 0), "signal"] = -1
```

```python
# Copy the new "signal" column to a new Series called `y`.
y = df_stock_data["signal"].copy()
```

```python
#Split the Data in Training and Test Sets
# Import required libraries
from pandas.tseries.offsets import DateOffset

# Select the start of the training period
training_begin = X.index.min() + DateOffset(months=1)

# Display the training begin date
print(training_begin)
```

```python
# Select the ending period for the training data with an offset of 18 months
training_end = X.index.min() + DateOffset(months=4)

# Display the training end date
print(training_end)
```

```python
# Generate the X_train and y_train DataFrames
X_train = X.loc[training_begin:training_end]
y_train = y.loc[training_begin:training_end]
```

```python
# Generate the X_test and y_test DataFrames
X_test = X.loc[training_end:]
y_test = y.loc[training_end:]
```

```python
#Standardize the Data
# Import required libraries
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler instance
scaler = StandardScaler()
 
# Apply the scaler model to fit the X-train data
X_scaler = scaler.fit(X_train)
 
# Transform the X_train and X_test DataFrames using the X_scaler
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```
